```python
import pandas as pd
import numpy as np

# 👀 check available columns
print("Available columns:", df.columns.tolist())

# ---------------------------------------
# 1. Handle timestamp (adapt based on your dataset)
# ---------------------------------------
# Case 1: If dataset has 'Time' column (like Kaggle credit card fraud dataset: seconds since start)
if 'Time' in df.columns and 'timestamp' not in df.columns:
    df['timestamp'] = pd.to_datetime(df['Time'], unit='s')

# Case 2: If dataset already has datetime column (e.g., 'trans_date', 'trans_time', etc.)
elif 'trans_date' in df.columns and 'trans_time' in df.columns:
    df['timestamp'] = pd.to_datetime(df['trans_date'] + ' ' + df['trans_time'])

# Case 3: If already has 'TransactionDT'
elif 'TransactionDT' in df.columns:
    df['timestamp'] = pd.to_datetime(df['TransactionDT'], unit='s', origin='unix')

# ---------------------------------------
# 2. Time-based features
# ---------------------------------------
if 'timestamp' in df.columns:
    df['hour'] = df['timestamp'].dt.hour
    df['day_of_week'] = df['timestamp'].dt.dayofweek
    df['is_weekend'] = df['day_of_week'].isin([5,6]).astype(int)

# ---------------------------------------
# 3. User behavior features (only if user_id available)
# ---------------------------------------
if 'user_id' in df.columns:
    # Avg spend per user
    df['user_avg_amount'] = df.groupby('user_id')['amount'].transform('mean')

    # Daily txn count
    df['date'] = df['timestamp'].dt.date
    df['user_txn_count_day'] = df.groupby(['user_id','date'])['amount'].transform('count')

    # Location change (if location column exists)
    if 'location' in df.columns:
        user_main_location = df.groupby('user_id')['location'].agg(lambda x: x.mode()[0])
        df = df.merge(user_main_location.rename('main_location'), on='user_id')
        df['location_change'] = (df['location'] != df['main_location']).astype(int)

# ---------------------------------------
# 4. Rolling aggregations (per user)
# ---------------------------------------
if 'user_id' in df.columns:
    df = df.sort_values(by=['user_id','timestamp'])
    df['rolling_avg_amount'] = df.groupby('user_id')['amount'].transform(lambda x: x.rolling(window=5, min_periods=1).mean())
    df['rolling_std_amount'] = df.groupby('user_id')['amount'].transform(lambda x: x.rolling(window=5, min_periods=1).std())

# ---------------------------------------
# 5. Risk-based features (if fraud labels exist)
# ---------------------------------------
if 'is_fraud' in df.columns:
    if 'merchant_id' in df.columns:
        merchant_risk = df.groupby('merchant_id')['is_fraud'].mean()
        df = df.merge(merchant_risk.rename('merchant_risk_score'), on='merchant_id')

    if 'device_id' in df.columns:
        device_risk = df.groupby('device_id')['is_fraud'].mean()
        df = df.merge(device_risk.rename('device_risk_score'), on='device_id')

# ✅ Final check
print(df.head())
```

```python
# 📌 Imbalanced Classification with Advanced Techniques
# Includes: SMOTE, ADASYN, SMOTE-ENN, SMOTE-Tomek, and Deep Learning with Focal Loss

# Install required packages for imbalance handling
!pip install imbalanced-learn

# Imports
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.combine import SMOTEENN, SMOTETomek

# TensorFlow
import tensorflow as tf

# ----------------------------
# Custom Focal Loss (no addons)
# ----------------------------
def focal_loss(gamma=2., alpha=0.25):
    def focal_loss_fixed(y_true, y_pred):
        y_true = tf.cast(y_true, tf.float32)
        y_pred = tf.cast(y_pred, tf.float32)
        bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
        bce_exp = tf.math.exp(-bce)
        focal_loss_val = alpha * tf.pow((1 - bce_exp), gamma) * bce
        return focal_loss_val
    return focal_loss_fixed

# ----------------------------
# Create Synthetic Imbalanced Dataset
# ----------------------------
X, y = make_classification(n_samples=5000, n_features=20, n_informative=2,
                           n_redundant=10, n_clusters_per_class=1,
                           weights=[0.95], flip_y=0, random_state=42)

print("Original dataset shape:", pd.Series(y).value_counts())

# ----------------------------
# Resampling Methods
# ----------------------------
samplers = {
    "SMOTE": SMOTE(),
    "ADASYN": ADASYN(),
    "SMOTE-ENN": SMOTEENN(),
    "SMOTE-Tomek": SMOTETomek()
}

for name, sampler in samplers.items():
    X_res, y_res = sampler.fit_resample(X, y)
    print(f"{name} resampled shape:", pd.Series(y_res).value_counts())

# ----------------------------
# Train-Test Split + Scaling
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ----------------------------
# Logistic Regression (Baseline)
# ----------------------------
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("\nLogistic Regression Report:\n", classification_report(y_test, y_pred))

# ----------------------------
# Deep Learning with Focal Loss
# ----------------------------
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

model.compile(optimizer="adam",
              loss=focal_loss(alpha=0.25, gamma=2),
              metrics=["accuracy", tf.keras.metrics.AUC(name="auc")])

history = model.fit(X_train, y_train, validation_data=(X_test, y_test),
                    epochs=10, batch_size=32, verbose=1)

# Evaluate
y_pred_nn = (model.predict(X_test) > 0.5).astype("int32")
print("\nNeural Network with Focal Loss Report:\n", classification_report(y_test, y_pred_nn))
```

```python
# =============================
# Fraud Detection Pipeline
# =============================

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.metrics import classification_report, roc_auc_score
import xgboost as xgb
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam

# -------------------------------
# Simulated Transaction Data
# -------------------------------
np.random.seed(42)
n_samples = 2000
X = np.random.rand(n_samples, 10)
y = np.random.randint(0, 2, n_samples)

# -------------------------------
# Train/Test Split
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# -------------------------------
# Autoencoder for Feature Learning
# -------------------------------
input_dim = X_train.shape[1]
encoding_dim = 5

input_layer = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation="relu")(input_layer)
decoded = Dense(input_dim, activation="sigmoid")(encoded)

autoencoder = Model(input_layer, decoded)
encoder = Model(input_layer, encoded)

autoencoder.compile(optimizer=Adam(0.01), loss="mse")
autoencoder.fit(
    X_train, X_train,
    epochs=20, batch_size=32,
    shuffle=True, validation_split=0.2,
    verbose=0
)

X_train_encoded = encoder.predict(X_train)
X_test_encoded = encoder.predict(X_test)

# -------------------------------
# Base Models for Stacking
# -------------------------------
rf = RandomForestClassifier(n_estimators=100, random_state=42)
xgb_clf = xgb.XGBClassifier(use_label_encoder=False, eval_metric="logloss")
meta_learner = RandomForestClassifier(n_estimators=50, random_state=42)

stacking_model = StackingClassifier(
    estimators=[("rf", rf), ("xgb", xgb_clf)],
    final_estimator=meta_learner,
    cv=5
)

# Train Stacking Model
stacking_model.fit(X_train_encoded, y_train)
y_pred = stacking_model.predict(X_test_encoded)
print("=== Stacking Model Report ===")
print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_pred))

# -------------------------------
# Semi-Supervised Learning (Pseudo-labeling)
# -------------------------------
unlabeled_data = np.random.rand(500, 10)
unlabeled_encoded = encoder.predict(unlabeled_data)

pseudo_labels = stacking_model.predict(unlabeled_encoded)
X_train_semi = np.vstack([X_train_encoded, unlabeled_encoded])
y_train_semi = np.hstack([y_train, pseudo_labels])

stacking_model.fit(X_train_semi, y_train_semi)
y_pred_semi = stacking_model.predict(X_test_encoded)
print("\n=== Semi-Supervised Report ===")
print(classification_report(y_test, y_pred_semi))
print("ROC-AUC:", roc_auc_score(y_test, y_pred_semi))

# -------------------------------
# Graph-Based Fraud Detection
# -------------------------------
import networkx as nx

# Create synthetic graph (Users ↔ Merchants)
G = nx.Graph()
users = [f"user_{i}" for i in range(50)]
merchants = [f"merchant_{j}" for j in range(20)]

# Add edges (simulate transactions)
for u in users:
    for _ in range(np.random.randint(1, 5)):
        m = np.random.choice(merchants)
        amount = np.random.rand() * 1000
        G.add_edge(u, m, weight=amount)

# ✅ Updated graph info (since nx.info is removed in NetworkX >= 3.0)
print("\nGraph Info:")
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
print("Is Directed:", G.is_directed())

# Simple Graph Feature: Degree Centrality
degree_centrality = nx.degree_centrality(G)
sorted_centrality = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)

print("\nTop Suspicious Accounts (high centrality):")
for node, score in sorted_centrality[:10]:
    print(node, "->", round(score, 3))
```

```python
# =============================
# Fraud Detection Pipeline with Evaluation & Deployment
# =============================

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    classification_report, roc_auc_score,
    precision_recall_curve, auc, f1_score
)
from sklearn.ensemble import IsolationForest
import xgboost as xgb
from imblearn.over_sampling import SMOTE
import shap
import matplotlib.pyplot as plt

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam

# -------------------------------
# Simulated Transaction Dataset
# -------------------------------
np.random.seed(42)
n_samples = 3000
data = pd.DataFrame({
    "amount": np.random.rand(n_samples) * 500,
    "merchant": np.random.choice(["A", "B", "C", "D"], n_samples),
    "category": np.random.choice(["food", "tech", "fashion"], n_samples),
    "timestamp": pd.date_range("2023-01-01", periods=n_samples, freq="h"),
    "fraud": np.random.randint(0, 2, n_samples)
})

# -------------------------------
# Feature Engineering for Datetime
# -------------------------------
data["hour"] = data["timestamp"].dt.hour
data["dayofweek"] = data["timestamp"].dt.dayofweek
data = data.drop(columns=["timestamp"])

# -------------------------------
# Train/Test Split
# -------------------------------
X = data.drop(columns=["fraud"])
y = data["fraud"]

X_train_df, X_test_df, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# -------------------------------
# Preprocessing (Scaling + OneHot)
# -------------------------------
numeric_features = ["amount", "hour", "dayofweek"]
categorical_features = ["merchant", "category"]

numeric_transformer = Pipeline(steps=[("scaler", StandardScaler())])
categorical_transformer = Pipeline(steps=[("encoder", OneHotEncoder(handle_unknown="ignore"))])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features),
    ]
)

# Fit + transform
X_train = preprocess.fit_transform(X_train_df)
X_test = preprocess.transform(X_test_df)

# -------------------------------
# Handle Imbalance with SMOTE
# -------------------------------
smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)

# -------------------------------
# Autoencoder for Feature Learning
# -------------------------------
input_dim = X_train.shape[1]
encoding_dim = 8

inp = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation="relu")(inp)
decoded = Dense(input_dim, activation="sigmoid")(encoded)

autoencoder = Model(inp, decoded)
encoder = Model(inp, encoded)

autoencoder.compile(optimizer=Adam(0.01), loss="mse")
autoencoder.fit(X_train, X_train, epochs=15, batch_size=64,
                shuffle=True, validation_split=0.2, verbose=0)

X_train_encoded = encoder.predict(X_train)
X_test_encoded = encoder.predict(X_test)

# -------------------------------
# Models: Isolation Forest + XGBoost
# -------------------------------
iso = IsolationForest(contamination=0.1, random_state=42)
iso.fit(X_train_encoded)
y_pred_iso = (iso.predict(X_test_encoded) == -1).astype(int)

xgb_clf = xgb.XGBClassifier(eval_metric="logloss", use_label_encoder=False)
xgb_clf.fit(X_train_encoded, y_train)
y_pred_xgb = xgb_clf.predict(X_test_encoded)
y_proba_xgb = xgb_clf.predict_proba(X_test_encoded)[:, 1]

# -------------------------------
# Evaluation Metrics
# -------------------------------
print("\n=== Isolation Forest Report ===")
print(classification_report(y_test, y_pred_iso))

print("\n=== XGBoost Report ===")
print(classification_report(y_test, y_pred_xgb))
print("ROC-AUC:", roc_auc_score(y_test, y_proba_xgb))

# Precision-Recall Curve
precision, recall, thresholds = precision_recall_curve(y_test, y_proba_xgb)
pr_auc = auc(recall, precision)
print("PR-AUC:", pr_auc)

plt.plot(recall, precision, marker='.')
plt.title("Precision-Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.show()

# -------------------------------
# Threshold Tuning
# -------------------------------
best_thresh = thresholds[np.argmax(2*precision*recall/(precision+recall+1e-6))]
print("\nBest threshold:", best_thresh)

y_pred_thresh = (y_proba_xgb >= best_thresh).astype(int)
print("\n=== Tuned Threshold Report ===")
print(classification_report(y_test, y_pred_thresh))
print("F1:", f1_score(y_test, y_pred_thresh))

# -------------------------------
# Explainability with SHAP
# -------------------------------
explainer = shap.TreeExplainer(xgb_clf)
shap_values = explainer.shap_values(X_test_encoded[:200])

print("\nShowing SHAP summary plot...")
shap.summary_plot(shap_values, X_test_encoded[:200], show=False)
plt.show()
```
